```python
flips = np.random.choice(["H", "T"], 10)
flips
```

```python
np.count_nonzero(flips == "T") / 10.0
```

```python
flips = np.random.choice(["H", "T"], 100)
np.count_nonzero(flips == "T") / 100.0
```

```python
flips = np.random.choice(["H", "T"], 1000)
np.count_nonzero(flips == "T") / 1000.0
```

```python
flips = np.random.choice(["H", "T"], 1000000000)
np.count_nonzero(flips == "T") / 1000000000.0
```
